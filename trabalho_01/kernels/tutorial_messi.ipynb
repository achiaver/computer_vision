# Tutorial do Messi

Abrir, Mostrar e Salvar uma imagem usando a biblioteca opencv.

## Bibliotecas relevantes

```python
import numpy as np
import cv2

import matplotlib.pyplot as plt

%matplotlib inline
```

## Abrindo uma imagem

```python
img = cv2.imread('../images/opencv_messi.jpg', 0)
```

## Mostrando uma imagem

```python
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([])
plt.yticks([])
plt.show()
```

## Salvando uma image

```python
cv2.imwrite('../images/messi_2.png', img)
```

```python
img = cv2.imread('../images/messi_2.png', 0)

plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([])
plt.yticks([])
plt.show()
```
