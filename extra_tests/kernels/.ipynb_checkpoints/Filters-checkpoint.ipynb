```python
import numpy as np
import cv2 as cv

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img1 = cv.imread('../images/IMG_1114.JPG', 1)
img2 = cv.imread('../images/IMG_1115.JPG', 1)
```

```python
fig = plt.figure(figsize = (15, 7))

ax1 = plt.subplot(1, 2, 1)
plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
plt.axis('off')

ax2 = plt.subplot(1, 2, 2)
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
print(img1.shape)
print(img2.shape)
```

```python
img1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
img2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
```

```python
fig = plt.figure(figsize = (15, 7))

ax1 = plt.subplot(1, 2, 1)
plt.imshow(img1, cmap = 'gray')
plt.axis('off')

ax2 = plt.subplot(1, 2, 2)
plt.imshow(img2, cmap = 'gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
def histograma(img):
    m, n = img.shape  
    hist, bins = np.histogram(img, range = (0, 256), bins = 64)
    hist = hist / float(m * n)
    center = (bins[:-1] + bins[1:]) / 2
    mean = np.mean(img)
    std = np.std(img)
        
    return hist, center, mean, std
```

```python
hist = histograma(img1)
```

```python
fig = plt.figure(figsize = (12, 5))

m_y = 0.005

ax1 = plt.subplot(1, 2, 1)
plt.bar(hist[1], hist[0], align = 'center', color = 'm', width = 4)
plt.xlabel('Intensidade', fontsize = 18)
plt.ylabel('$p(f(x, y))$', fontsize = 18)
plt.plot(hist[2], m_y, 'ko')
plt.plot([hist[2] - hist[3], hist[2] + hist[3]], [m_y] * 2, 'k--')
plt.title('IMG_1')
plt.xlim([0, 256])
plt.ylim([0, 0.08])

plt.show()
```
